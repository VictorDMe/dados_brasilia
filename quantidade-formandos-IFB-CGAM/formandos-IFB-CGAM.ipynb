```python
import csv
import matplotlib.pyplot as plt
```

```python
plt.plot(lista_anos, contador_certificados,
         color='red', 
         marker='.', 
         linestyle='solid')

plt.title('Quantidade de certificados emitidos pelo IFB CGAM')
plt.ylabel('# de certificados')
plt.grid()
plt.draw()
plt.axis([2012, 2019, 0, 400])
plt.savefig('grafico-linhas', dpi=150)
plt.show()
```

```python
zip_anos_certificados = list(zip(lista_anos, contador_certificados))
colunas = ['Ano', 'Quan. de certificados']
bbox = [0,0,1,1]

fig, ax = plt.subplots(1,1)
ax.table(cellText=zip_anos_certificados, colColours=['yellow'] * 2,
         colLabels=colunas, cellLoc='center',
         colWidths=[0.5 for x in colunas],
         bbox=bbox, colLoc='center')

ax.axis('tight')
ax.axis('off')
plt.title('Quantidade de certificados emitidos pelo IFB CGAM')
plt.savefig('tabela', dpi=150)
plt.show()
```
