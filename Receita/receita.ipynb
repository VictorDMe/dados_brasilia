```python
import matplotlib.pyplot as plt
import numpy as np
import locale

locale.setlocale(locale.LC_MONETARY, 'pt_BR.UTF-8')
```

```python
tax = [('IPVA', 1207235461.35),
            ('ISS', 1784945872.81),
            ('ICMS', 7938331061.22),
            ('IPTU', 1166478560.76),
            ('ITBI', 559906507.37),
            ('ITCD', 202040439.66),
            ('IR', 2709444348.69)]
tax.sort(key=lambda value: value[1])

total = 0
for _, valor in tax:
    total += valor

total = round(total, 2)
```

```python
tax_names = np.array([f'{name} ({round(value * 100 / total, 2)}%)'
                     for (name, value) in tax])

tax_values = np.array([value for _, value in tax])
plt.title('Porcentagem de arrecadação por imposto no DF')
plt.pie(tax_values, labels=tax_names, shadow=True,
       )

#plt.figure(figsize=(8,6), dpi=150, facecolor='white')
plt.savefig('pie_graph', transparent=False, dpi=200)
plt.show()
```
