```python
import matplotlib.pyplot as plt
import numpy as np
import locale

locale.setlocale(locale.LC_MONETARY, 'pt_BR.UTF-8')
```

## Atribuindo os valores de cada imposto e obtendo o custo total

```python
tax = [('IPVA', 1207235461.35),
            ('ISS', 1784945872.81),
            ('ICMS', 7938331061.22),
            ('IPTU', 1166478560.76),
            ('ITBI', 559906507.37),
            ('ITCD', 202040439.66),
            ('IR', 2709444348.69)]
tax.sort(key=lambda value: value[1])

total = 0
for _, value in tax:
    total += value

total = round(total, 2)
```

## Criando gr√°fico de pizza

```python
tax_names = np.array([f'{name} ({round(value * 100 / total, 2)}%)'
                     for (name, value) in tax])

plt.title('Porcentagem de arrecada√ß√£o por imposto no DF')
plt.pie([value for _, value in tax], 
        labels=tax_names, shadow=True,)
plt.savefig('pie_graph', dpi=200)
plt.show()
```

## Criando tabela

```python

format_tax = list(zip([tax_name for tax_name, _ in tax],
                      [locale.currency(tax_value, grouping=True) 
                       for _, tax_value in tax]))
fig, ax = plt.subplots()
fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')

plt.table(cellText=format_tax,
          colLabels=['Imposto', 'Arrecada√ß√£o'],
          loc='center')
fig.tight_layout()
plt.savefig('table_graph', dpi=200)
plt.show()
```
